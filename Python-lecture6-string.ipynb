```python
# string operation.
## concatenation, 加法，或者合并
str6= str1+str2
print(str6)

str6= str1+str2+str(3)
print(str6)
print("".join([str1,str2]))
print("\t".join([str1,str2]))
print("\n".join([str1,str2]))
```

```python
# string multiplication
print("I love python!\n"*99)
```

```python
# string slicing
stra= 'banananananananal'
print(stra[0])
print(stra[1])
# slicing method: [start,end,step)
print(stra[0:5])
print(stra[0:6])
print(stra[:6])
print(stra[1:])
print(stra[1::2])
print(stra[::2])
print(stra[-1])
print(stra[-2])
print(stra[::-1])
print("i love python"[::-1])
a= "nohtyp evol i"
print(a[::-1])








```
